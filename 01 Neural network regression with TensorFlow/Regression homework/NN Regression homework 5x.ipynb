# Homework 1
Write a neural network that takes a single numerical input and predicts a single numerical output.

```python
import tensorflow as tf
import numpy as np
tf.__version__
```

```python
X_ = np.arange(1, 20, 1, dtype="float32")
y = np.pow(X_,2)
X_,y
```

```python
X_ = tf.constant(X_)
X = tf.expand_dims(X_, axis=-1)
y = tf.constant(y)
print(X)
print(y)
```

```python
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(15, activation="relu"),
    tf.keras.layers.Dense(25, activation="relu"),
    tf.keras.layers.Dense(1)
])


model_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
                loss=tf.keras.losses.mse,
                metrics=["mae"])


model_1.fit(X,y,epochs=155)
```

```python
pred = tf.constant([1.,2.,5.,20])
model_1.predict(pred)
```

# Homework 2
Example: House square footage and number of rooms в†’ price estimate. You can create your own data, and the model can learn both features together.

```python
import tensorflow as tf
import numpy as np

tf.__version__
```

```python
home_X = tf.constant([[70,2],
                      [80,2],
                      [100,3],
                      [50,1],
                      [150,4],
                      [80,2],
                      [80,3],
                      [110,4],
                      [90,2],
                      [100,4]], dtype="float32")



# fiyatlar arasД±nda korelasyon oluЕџturalД±m
price_list_y = []
price_y = tf.constant([])

for i in home_X:
    price_glitch = np.random.random(1)*5000
    price = np.array(i[0]*10000) + np.array(i[1]*15000) + np.array(price_glitch)
    price_list_y.append(price)

price_y = tf.constant(price_list_y)


def calculate_y_ground(home_X):
    _price_glitch = 5000
    _price = np.array(home_X[0][0]*10000) + np.array(home_X[0][1]*15000) + np.array(_price_glitch)
    return _price


print(price_y)
print(calculate_y_ground([[70,2]]))
```

```python
home_X
```

```python
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(15),
    tf.keras.layers.Dense(20, activation="relu"),
    tf.keras.layers.Dense(25, activation="relu"),
    tf.keras.layers.Dense(1)
])


model_2.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
    loss = tf.keras.losses.MSE,
    metrics = ["mae"]
)

model_2.fit(home_X, price_y, epochs=150)
```

```python
pred = tf.constant([[70. ,3. ],[80., 3.]])

print("X: \n", pred)
print("\nExpected y: \n", calculate_y_ground(pred))


print("\npredicted y:")
model_2.predict(pred)
```

# Homework 3
Use real dataset

```python
import tensorflow as tf
import numpy as np

tf.__version__
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
data = fetch_california_housing()

print(data.data)
print(data.target)
print(len(data.data)) #20640 sample

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# daha derin bi model oluЕџturalД±m

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(15),
    tf.keras.layers.Dense(30, activation="relu"),
    tf.keras.layers.Dense(50, activation="relu"),
    tf.keras.layers.Dense(1),
])

model_3.compile(
    optimizer = tf.keras.optimizers.Adam(),
    loss= tf.keras.losses.MSE,
    metrics = ["mae"]
)

callback = tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
model_3.fit(X_train, y_train, epochs=200, callbacks=[callback], validation_data=(X_test, y_test))

model_3.evaluate(X_test, y_test)
```

# Homework 5
Transform what you've done so far into a Streamlit application. Users enter square footage and number of rooms, the model makes an estimate, and the price appears on the screen. Upload and use the model you've saved.

**Note! Streamlit doesn't work with .ipynb files; use it in a .py file.**

```python
import tensorflow as tf
import streamlit as st
import os
```

```python
path = os.path.join(os.getcwd(),"Regression homework","model_2.keras")

model = tf.keras.models.load_model(path)

st.title("House Price Predictor")

sq_meters = st.text_input("Metrekare girin", 80)

n_of_room = st.text_input("oda sayД±sД± girin", 3)

if sq_meters and n_of_room:
    if not sq_meters.isdigit() or not n_of_room.isdigit():
        st.error("Enter a number")

    sq_meters = float(sq_meters)
    n_of_room = float(n_of_room)
    pred = tf.constant([[sq_meters, n_of_room]])
    st.markdown("---")
    st.markdown("**Price:**")
    st.text(f"$ {model.predict(pred)[0][0]:,.0f}")
```
